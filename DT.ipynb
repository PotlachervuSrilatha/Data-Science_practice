```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_excel("/content/Cricket_dt.xlsx")
df
```

```python
LE = LabelEncoder()
df["Gender"] = LE.fit_transform(df["Gender"])
df["Class"] = LE.fit_transform(df["Class"])
df["Cricket"] = LE.fit_transform(df["Cricket"])
```

```python
df.head()
```

```python
X=df[["Gender","Class"]]
Y=df["Cricket"]
```

```python
DT=DecisionTreeClassifier()
DT.fit(X,Y)
y_pred=DT.predict(X)
```

```python
ac = accuracy_score(Y,y_pred)
print("Accuracy Score: ",ac.round(3))
```

```python
from sklearn import tree
import graphviz
dot_data=tree.export_graphviz(DT, filled=True, rounded=True,
                              special_characters=True)
graph=graphviz.Source(dot_data)
graph
```

```python
print("Number of Nodes", DT.tree_.node_count)
print("Number of Depth", DT.tree_.max_depth)
```
