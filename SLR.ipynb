```python
import numpy as np
age = np.array([[24],[28],[30],[34],[36],[40]])
weight= np.array([70,74,75,83,88,90])
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,weight)
plt.show()
```

```python
x = age
y = weight
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x,y)
```

```python
lr.intercept_  # constant/intercept
```

```python
lr.coef_ # slope
```

```python
# manual calc
35.809523809523824 + (1.38095238*25)
```

```python
# machine calc
lr.predict(np.array([[25]]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
age = np.array([24, 28, 30, 34, 36, 40])
weight = np.array([70, 74, 75, 83, 88, 90])

# Fit a straight line
coefficients = np.polyfit(age, weight, 1)
poly = np.poly1d(coefficients)

# Plotting
plt.scatter(age, weight, color='blue', label='Data Points')
plt.plot(age, poly(age), color='red', label='Straight Line Fit')

# Labels and Legend
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age vs Weight with Straight Line Fit')
plt.legend()

# Show plot
plt.show()

```
